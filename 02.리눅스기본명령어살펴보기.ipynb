# man
명령어 커맨드의 메뉴얼(사용법)을 살펴 볼 수 있는 명령어.  
메뉴얼 정보는 /usr/share/man 해당 위치에 파일로써 존재하며, man 명령어 호출시 저장 섹션과 내용을 출력 하여 준다.



**기본형식**
```
man <options> keyword
```
    
**옵션**
- -k : 메뉴얼 목록을 검색
- -s : section-number 입력한 섹셕에 대한 메뉴얼을 출력
  - man1 : User Commands
  - man2 : System Calls
  - man3 : Sub routines
  - man4 : Devices
  - man5 : File Formats
- space : next page
- enter : nest line
- b : backward
- quit : quit

````python
# 명령어별 메뉴얼 정보가 저장 되어 있는 폴더
!ls /usr/share/man
````

````python
!man ls
````

````python
!man -k 1 | grep ls
````

# ls
대상 위치의 파일 목록을 출력 한다.

**기본형식**
```
ls <options> <file|directory>
```

**옵션**
- -a: .bashrc 와 같이 숨겨진 파일들 까지 모두 출력하는 옵션
- -l: 파일 및 디렉토리의 자세한 정보를 출력하는 옵션 (type, permission, link, size, owner)
- -R: 하위 디렉토리 까지 모두 포함하여 출력
- -d: 디렉토리가 내용이 아닌 디렉토리 자체를 출력


````python
!ls -a
````

````python
!ls -al
````

````python
!ls -alR
````

````python
# 디렉토리의 속성 정보를 출력
!ls -ld /tmp
````

# mkdir, rmdir, cd


### mkdir : 디렉토리를 생성 한다.

**기본형식**
```
mkdir <options> <directory_name>
```

**옵션**
- -m : 퍼미션을 설정한다.
- -p : 부모 디렉토리를 포함 하여 생성



### rmdir : 디렉토리를 삭제한다. 

**기본형식**
```
rmdir <options> <directory_name>
```

**옵션**
- -p : (비어있는) 부모 디렉토리를 포함하여 삭제


### cd : 현재 디렉토리 위치를 변경한다. 
**기본형식**
```
cd <directory_name>
```

**옵션**
- ~ : 홈 디렉토리로 이동
- \- : 디렉토리 이동을 하기 이전 디렉토리로 이동

````python
# 현재위치 확인
!pwd
````

````python
# sample 디렉토리 생성
!mkdir sample
````

````python
# 목록 생성
!ls
````

````python
# 부모 디렉토리를 포함하여 생성
!mkdir -p ./sample/parent/child
!mkdir -p -m 777 ./sample/parent/permission
````

````python
# 생성 정보 확인
!ls -al ./sample/parent
````

````python
# 부모 디렉토리를 포함하여 삭제
!rmdir sample/parent/permission
!rmdir -p sample/parent/child
!ls
````

# cp
파일을 복사 한다. 리눅스 시스템은 기본적으로 덮어쓰기가 활성화 되어 있다.

**기본형식**
```
cp <options> <origin_file_name> <target_file_name>
```

**옵션**
- -i : 복사시 overwrite 여부를 묻는다. (interactive)
- -f : 복사시 overwrite 여부 확인 없이 덮어쓴다. (force)
- -r : 디렉토리를 복사 한다. (recursive)

````python
# 임시 파일 생성
!touch ./sample/parent/tempfile
!ls ./sample/parent
````

````python
# 임시파일 복사
# !rm -rf tempfile -- 기존파일 삭제
!cp ./sample/parent/tempfile ./sample/parent/copyfile 
!ls ./sample/parent/
````

````python
# /etc/hosts 파일을 현재 디렉토리로 복사
!ls /etc/hosts
!cp /etc/hosts .
!ls 
````

````python
# 복사된 파일 확인
!ls
````

````python
# 복수개의 파일을 하나의 디렉토리로 복사
!cp /etc/hosts /etc/passwd ./sample/
````

````python
!arch -arm64 brew install tree
!tree
````

````python
# 폴더 복사 -r
!cp -r /etc/hosts ./sample/parent/
!ls ./sample/parent/
````

# mv
파일을 이동 한다.

**기본형식**
```
mv <options> <origin_file_name> <new_file_name>
```


**옵션**
- -i : overwrite 여부를 묻는 옵션
- -f : overwrite 여부를 묻지 않고 강제 덮어 쓰기
